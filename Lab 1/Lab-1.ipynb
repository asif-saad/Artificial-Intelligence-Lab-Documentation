# 1. Basic Python Programming
##a. Syntax and Semantics
##Python is known for its simple, readable syntax. The code structure uses indentation instead of curly braces or keywords.

````python
# This is a comment
# Python is case-sensitive
x = 5  # Variable assignment
X=0
print(x)  # Output: 5
print(X)
````

##b. Data Types
##Python supports multiple data types such as integers, floats, strings, and booleans.##

````python
# Integer
x = 5
print(type(x))  # <class 'int'>

# Float
y = 5.5
print(type(y))  # <class 'float'>

# String
name = "John"
print(type(name))  # <class 'str'>

# Boolean
is_valid = True
print(type(is_valid))  # <class 'bool'>
````

##c. Variables and Assignments
##Variables are assigned values using the = operator. The type of the variable is inferred automatically.

````python
x = 10
y = 20
z = x + y
print(z)  # Output: 30
````

##d. Conditional Statements
##Python provides ```if```, ```else```, and ```elif``` for conditional branching.

````python
x = 10
if (x > 5):
    print("x is greater than 5")
elif x == 5:
    print("x is 5")
else:
    print("x is less than 5")
````

##e. Loops
##Python supports two types of loops: `for` and `while`.

````python
# For loop with range
for i in range(1,5,2):
    print(i)  # Output: 0 1 2 3 4


# While loop
i = 0
while i < 5:
    print(i)  # Output: 0 1 2 3 4
    i += 1
````

##f. Functions
##Functions in Python are defined using the `def` keyword.

````python
def add(a, b):
    return a + b

result = add(5, 10)
print(result)  # Output: 15
````

##g. Exception Handling
##Python uses `try`, `except`, and `finally` for exception handling.

````python
try:
    x = 10 / 1
except ZeroDivisionError:
    print("Cannot divide by zero!")
finally:
    print("This will always be executed.")
````

#3. Object-Oriented Programming (OOP)
##a. Classes and Objects
##A class is a blueprint for creating objects (instances).

````python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def display_info(self):
        print(f"{self.year} {self.make} {self.model}")

# Creating an object
my_car = Car("Toyota", "Corolla", 2020)
my_car.display_info()  # Output: 2020 Toyota Corolla
````

##b. Inheritance
##Inheritance allows a class to inherit methods and properties from another class.

````python
class ElectricCar(Car):
    def __init__(self, make, model, year, battery_size):
        super().__init__(make, model, year)
        self.battery_size = battery_size

    def display_battery(self):
        print(f"This car has a {self.battery_size}-kWh battery.")

# Creating an object of the subclass
my_electric_car = ElectricCar("Tesla", "Model S", 2022, 100)
my_electric_car.display_info()  # Output: 2022 Tesla Model S
my_electric_car.display_battery()  # Output: This car has a 100-kWh battery.
````

##c. Polymorphism
##Polymorphism allows methods to have different implementations depending on the object.

````python
class Dog:
    def sound(self):
        return "Bark"

class Cat:
    def sound(self):
        return "Meow"

def make_sound(animal):
    print(animal.sound())

make_sound(Dog())  # Output: Bark
make_sound(Cat())  # Output: Meow
````

##d. Encapsulation
##Encapsulation involves hiding the internal state and requiring all interactions to be performed through methods.

````python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def get_name(self):
        return self.__name

    def set_age(self, age):
        self.__age = age

# Creating an object
person = Person("Alice", 30)
print(person.get_name())  # Output: Alice
person.set_age(31)
````

#6. Iterators and Generators
##a. Iterators

````python
# Creating an iterator from a list
my_list = [1, 2, 3]
iterator = iter(my_list)
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
````

##b. Generators

````python
# Generator function
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

gen = count_up_to(3)
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
````

#7. Decorators and Lambda Functions
##a. Decorators
A decorator is a function that modifies the behavior of another function.

````python
def decorator(func):
    def wrapper():
        print("Something before the function call.")
        func()
        print("Something after the function call.")
    return wrapper

@decorator
def say_hello():
    print("Hello!")

say_hello()
````

##b. Lambda Functions
Lambda functions are small anonymous functions.

````python
# Lambda function to add two numbers
add = lambda x, y: x + y
print(add(5, 10))  # Output: 15
````

#8. Comprehensions
##a. List Comprehensions

````python
# Square of even numbers between 1 and 10
squares = [x**2 for x in range(1, 11) if x % 2 == 0]
print(squares)  # Output: [4, 16, 36, 64, 100]
````
